```python
survey_log_records = []
for idx,_ in survey_log.iterrows():
    values = []
    for j in range(len(survey_log['survey_log_record'][idx])):
        try: 
            del survey_log['survey_log_record'][idx][j]['etc']
            temp = survey_log['survey_log_record'][idx][j].values() ## get values of survey_log_record's list
            temp = list(temp)
            values.append(temp)
        except:
            temp = survey_log['survey_log_record'][idx][j].values() ## get values of survey_log_record's list
            temp = list(temp)
            values.append(temp)
    values = np.array(values)
    values = values.flatten()
    survey_log_records.append(values)
```
