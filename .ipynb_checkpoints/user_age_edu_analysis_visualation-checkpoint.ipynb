```python
# 4단계 : sql문 작성(데이터 read)
query = "SELECT gl.*, ur.*, o.organization_idx FROM sb_db.brain_game_log gl LEFT JOIN sb_db.map_assignment_unit_user mauu ON gl.assignment_unit_user = mauu.assignment_unit_user_idx JOIN sb_db.user_real ur ON mauu.user = ur.idx JOIN brain_organization  o ON ur.organization = o.organization_name"
# 5단계 : pd로 sql문 실행
game_log = pd.read_sql(query, db)
```
