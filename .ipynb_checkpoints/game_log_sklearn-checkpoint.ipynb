```python
import numpy as np
import pymysql
import pandas as pd

from tqdm import tqdm
import sklearn
```

```python
game_log = pd.read_csv("game_log_features2.csv", index_col="game_log_idx")
```

```python
game_log_age = pd.DataFrame(game_log["level"])
game_log_edu = pd.DataFrame(game_log["level2"])

del game_log["level"]
del game_log["level2"]
del game_log["user"]
```

```python
game_log_age.value_counts()
```

```python
from itertools import compress
game_age_nan = [int(n) for n in pd.isnull(game_log["level"])]
game_age_nan = list(compress(range(61545),game_age_nan))
```

```python
game_log.drop(index=game_log.iloc[game_age_nan].index.tolist(), inplace=True)
```

```python
game_log = game_log.fillna(0)
game_log
```

```python
from sklearn.preprocessing import RobustScaler

robustScaler = RobustScaler()
train_data_robustScaled = robustScaler.fit_transform(game_log)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
encoder = OneHotEncoder()
encoder2 = LabelEncoder()

game_age_1hot = encoder.fit_transform(game_log_age)
game_age_label = encoder2.fit_transform(game_log_age)
game_age_1hot
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(game_log, game_age_label, test_size=0.2, shuffle=True, stratify=game_age_label, random_state=34)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log_clf = LogisticRegression(solver="lbfgs", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=500, random_state=42)
svm_clf = SVC(gamma="scale", random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard')
```

```python
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```
