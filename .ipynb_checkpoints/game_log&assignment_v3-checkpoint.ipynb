```python
import numpy as np
import pymysql
import pandas as pd

from tqdm import tqdm
```

```python
# 4단계 : sql문 작성(데이터 read)
query = "SELECT * FROM sb_db.brain_game_log A LEFT JOIN sb_db.assignment_unit_user_info B ON A.assignment_unit_user = B.assignment_unit_user_idx;"
# 5단계 : pd로 sql문 실행
game_log = pd.read_sql(query, db)
```

```python
game_log.set_index('game_log_idx', inplace=True)
```

```python
temp  = np.unique(game_log["user"])
temp = temp[~np.isnan(temp)]
len(temp)
```

```python
for idx, val in tqdm(game_log.iterrows(), total = game_log.shape[0]):
    if np.isnan(val["user"]):
        game_log = game_log.drop(idx)
```

```python
for idx, val in tqdm(game_log.iterrows(), total=game_log.shape[0]):
    game_log['game_log_record'][idx] = eval(val['game_log_record'])
```

```python
game_log_records = []
n= []

for idx,_ in game_log.iterrows():
    values = []
    for j in range(len(game_log['game_log_record'][idx])):
        temp = game_log['game_log_record'][idx][j].values() ## get values of survey_log_record's list
        temp = list(temp)
        values.append(temp)
        
    values = np.array(values)
    values = values.flatten()
    game_log_records.append(values)
    
    n.append(len(values))
```
